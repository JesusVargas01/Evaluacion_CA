```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm


# Leer los datos del archivo de Excel
df = pd.read_excel('/content/es.xlsx')

# Extraer los datos de 
a = df['Altura']
t = df['Talla']
s = df['Sexo']


media_estatura = a.mean()
media_talla = t.mean()
std_estatura = a.std()
std_talla = t.std()



g = (1 / (std_estatura * np.sqrt(2 * 3.14159))) * np.exp(-((a - media_estatura)**2) / (2 * std_estatura**2))
plt.hist(a, bins=15)
plt.xlabel("estatura")
plt.ylabel("freq")
plt.show()
plt.scatter(a, g ,color='r')
plt.xlabel("estatura")
plt.ylabel("std_dev")
plt.show()

h = (1 / (std_talla * np.sqrt(2 * 3.14159))) * np.exp(-((t - media_talla)**2) / (2 * std_talla**2))
plt.hist(t, bins=15)
plt.xlabel("talla")
plt.ylabel("freq")
plt.show()
plt.scatter(t, h)
plt.xlabel("talla")
plt.ylabel("std_dev")
plt.show()

```

```python
#plantear las siguientes ecuaciones:
# 0.25A + 0.15B + 0C = 1.5 (Potasio)
# 0.45A + 0.5B + 0.75C = 5 (Nitrato)
# 0.3A + 0.35B + 0.25C = 3 (Fosfato)


import numpy as np

# Definir el sistema de ecuaciones
A = np.array([[0.25, 0.15, 0], [0.45, 0.5, 0.75], [0.3, 0.35, 0.25]])
b = np.array([1.5, 5, 3])

# Calcular la determinante de la matriz de coeficientes
detA = np.linalg.det(A)

# Calcular los demas determinantes
detA1 = np.linalg.det(np.column_stack((b, A[:,1], A[:,2])))
detA2 = np.linalg.det(np.column_stack((A[:,0], b, A[:,2])))
detA3 = np.linalg.det(np.column_stack((A[:,0], A[:,1], b)))

# Calcular los valores de las variables
x1 = detA1 / detA
x2 = detA2 / detA
x3 = detA3 / detA

# Imprimir la solucin
print("Las cantidades  son:")
print("A:", x1)
print("B:", x2)
print("C:", x3)
```

```python
#plantear las siguientes ecuaciones:
# 0.25A + 0.15B + 0C = 1.5 (Potasio)
# 0.45A + 0.5B + 0.75C = 5 (Nitrato)
# 0.3A + 0.35B + 0.25C = 3 (Fosfato)


import numpy

# Definir el sistema de ecuaciones
A_3 = np.array([[0.25, 0.15, 0], [0.45, 0.5, 0.75], [0.3, 0.35, 0.25]])
b_3 = np.array([1.5, 5, 3])
#Solucin
S= np.linalg.solve(A_3, b_3)

# Imprimir la solucin
print("Las cantidades  son:")
print("A:", S[0])
print("B:", S[1])
print("C:", S[2])
```
